```python
flights = sns.load_dataset("flights")
flights.head()
p_flights = flights.pivot(index = "year", columns = "month", values = "passengers")
```

```python
sns.distplot(flights.passengers.dropna()) #shows distribution of a number of passengers 
```

```python
sns.jointplot(data=flights, x="passengers", y="year", kind='reg', color='g') # the number of passengers increased over time probably due to advances in tech that allowed for greater airplanes
```

```python
sns.heatmap(p_flights, annot=True, vmin=50, vmax=600, fmt="d")
```

```python
ax = sns.lineplot(x='month', y='year', hue='passengers', data=flights)
```
